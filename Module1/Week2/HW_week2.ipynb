## EX1

```python
def ex1(num_list, k):
    n = len(num_list)
    for i in range(n-k+1):
        Maxx = -float('inf')
        for j in range(i,i+k):
            if num_list[j] > Maxx:
                Maxx = num_list[j]
        for j in range(i,i+k):
            if (num_list[j] < Maxx):
                num_list[j] = Maxx
    res_list = []
    for i in range(n-k+1):
        res_list.append(num_list[i])
    return res_list
```

```python
num_list = [3, 4, 5, 1, -44 , 5 ,10, 12 ,33, 1]
k = 3
print(ex1(num_list, k))
```

## EX2

```python
def ex2(string):
    n = len(string)
    res = {}
    for i in range(n):
        if (res.get(string[i]) is None):
            res[string[i]] = 0
        res[string[i]] += 1
    return res
```

```python
string = 'Happiness'
print(ex2(string))
```

## EX3


```python
file_path = 'P1_data.txt'
def ex3(file_path):
    with open(file_path, 'r') as f:
        content = f.read()
    words_list = content.lower().split()
    word_count = {} 
    for word in words_list:
        if word not in word_count:
            word_count[word] = 0
        word_count[word] += 1
    word_count =  dict(sorted(word_count.items()))
    return word_count
ex3(file_path)
```

# EX4

```python
def levenshtein_distance(source, target):
    rows = len(source) + 1
    cols = len(target) + 1
    
    a = [[0 for _ in range(cols)] for _ in range(rows)]
    
    for i in range(rows):
        a[i][0] = i
        
    for j in range(cols):
        a[0][j] = j
    for i in range(1,rows):
        for j in range(1,cols):
            x = a[i-1][j] + 1
            y = a[i][j-1] + 1
            if (source[i-1] == target[j-1]):
                z = a[i-1][j-1]
            else:
                z = a[i-1][j-1] + 1
            val = min(x,y,z)
            a[i][j] = val
    return a[rows-1][cols-1]
            
    
```

```python
levenshtein_distance('yu','you')
```

```python
levenshtein_distance('kitten','sitting')
```

```python
def my_function(data, max, min):

    result = []

    for i in data:
        #Your code here
        if (i<min):
            result.append(min)
        #Neu i < min thi them min vao result
        elif i > max:
            result.append(max)
        else:
            result.append(i)
    return result
my_list = [5, 2, 5, 0, 1]
max = 1
min = 0
assert my_function(max = max, min = min, data = my_list) == [1, 1, 1, 0, 1]
```
