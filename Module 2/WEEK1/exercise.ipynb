```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
img = mpimg.imread('./dog.jpeg')
imgplot = plt.imshow(img)
```

```python
import numpy as np
gray_img_01 = np.apply_along_axis(lambda x: (max(x[0],x[1],x[2])+min(x[0],x[1],x[2]))/2, 2, img)
gray_img_01[0, 0]
```

```python
def brighness(vector):
    return (max(vector[0],vector[1],vector[2]) + min(vector[0],vector[1],vector[2])) / 2
gray_img_02 = np.apply_along_axis(brighness, 2, img)
gray_img_02[0, 0]
```

```python
def average(vector):
    return (vector[0] + vector[1] + vector[2]) / 3
gray_img_03 = np.apply_along_axis(average, 2, img)
gray_img_03[0, 0]
```

```python
def luminosity(vector):
    return 0.21 * vector[0] + 0.72 * vector[1] + 0.07 * vector[2]
gray_img_04 = np.apply_along_axis(luminosity, 2, img)
gray_img_04[0, 0]
```
